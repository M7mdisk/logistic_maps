```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

n_iterations = 25
data = ["logistic","gauss","sine"]
from KC_methods import methods, datasets, calc_KC76
SAMPLES = 10**6
```

```python
for d in data:
    dff = pd.read_csv(f"logisitc_35699.csv")
    # dff = pd.read_csv(f"{d}_patterns.csv")

    patterns = list(dff["pattern"])
    probability = Counter(patterns)
    df = pd.DataFrame()
    df["x"] = patterns
    df["K"] = df["x"].apply(calc_KC76)
    df["scaled"] = len(patterns[0]) * ( df["K"]-df["K"].min() ) / ( df["K"].max()-df["K"].min() )
    fig = df["scaled"].hist(density=True)
    # plt.clf()
    plt.yscale("log")
    plt.ylabel("Frequency",fontsize=15)
    plt.xlabel(r'$\mathit{\tilde{K}(x)}$',fontsize=15)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.grid(False)
    plt.tight_layout()
    plt.savefig(f"hist/hist_scaled_logistic_35699.png")
    break
```
