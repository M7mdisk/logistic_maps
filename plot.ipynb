```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("slopes.csv")
df = df[df["slope"] <0]
```

```python
df["slop_norm"] = abs(np.log10(abs(df["slope"])))
df["slop_norm"]
```

```python

plt.scatter(df["r"],df["epsilon"],s=np.log10(df["slop_norm"])*1000,c=df["e_frequency"],alpha=0.5)
plt.xlabel("r")
c_bar = plt.colorbar()
c_bar.set_label("Kick frequency")
plt.ylabel("Kick size")
plt.savefig("start_1000.png")
```
