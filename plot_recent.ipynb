```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
# df = pd.read_csv("results/slopes.csv")
start_iter=1000
df = pd.read_csv("slopesssss_1000.csv")
df
df = df[df["slope"] < 0 ]
df
```

```python
df["slop_norm"] = abs(df["slope"])
df
```

```python
min_value = abs(df['slope']).min()
max_value = abs(df['slope']).max()

# Perform Min-Max scaling
df['scaled_s'] = (abs(df['slope']) - min_value) / (max_value - min_value)
```

```python
plt.clf()
plt.scatter(df["r"],df["epsilon"],s=df["scaled_s"]*500,c=df["e_frequency"],alpha=0.5)
plt.xlabel(r'$\mu$',fontsize=15)
c_bar = plt.colorbar()
c_bar.set_label(r'$q$',fontsize=15)
plt.ylabel(r'$\epsilon$',fontsize=15)
plt.savefig(f"start_{start_iter}.png")
```
