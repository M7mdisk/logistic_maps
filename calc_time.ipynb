```python
from KC_methods import methods
import random
import time
table = []
for method_name,method_func in methods.items():
    sum_mthd = 0
    for n in range(20,251,10):
        strs = ["".join([random.choice(["0", "1"]) for _ in range(n)]) for _ in range(500)]
        st = time.process_time()
        for seq in strs:
            method_func(seq)
        end = time.process_time()
        table.append([method_name,n,end-st])
        sum_mthd+= end- st

```

```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame(table,columns=["Method","n","Time"])
df["Method"] = df["Method"].str.replace("PLT","Plotnik").replace("LZ76_adapted","Adapted LZ67").replace("ENTROPY","Entropy")
import matplotlib.pyplot as plt

ax = sns.lineplot(data=df, x="n", y="Time", hue="Method")
plt.yscale("log")
plt.ylabel("Log(Time)")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```
