```python
samples = 1000000

import random
import numpy as np


ar = [f'{random.getrandbits(25):=025b}' for i in range(samples)]
```

```python
print(len(ar))

from KC_methods import calc_KC76

complexities = np.array([calc_KC76(a) for a in ar])



scaled = 25 * ( complexities-complexities.min() ) / ( complexities.max()-complexities.min() )
```

```python
import matplotlib.pyplot as plt

plt.hist(scaled
         ,density=True)
plt.yscale("log")
plt.ylabel("Frequency",fontsize=15)
plt.xlabel(r'$\mathit{\tilde{K}(x)}$',fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(False)
plt.tight_layout()

plt.savefig(f"distribution_random_n_25_1_mil.png")
```
